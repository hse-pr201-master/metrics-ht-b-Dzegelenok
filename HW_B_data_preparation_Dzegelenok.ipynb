```python
df["GDP growth rate (annual %, const. 2005 prices)"].unique()
```

Видно, что одно из значений записано неверно, поэтому приведем это значение к численному виду

```python
df["GDP growth rate (annual %, const. 2005 prices)"] = df["GDP growth rate (annual %, const. 2005 prices)"].replace("-~0.0", "0.0")
```

```python
df["GDP growth rate (annual %, const. 2005 prices)"] = df["GDP growth rate (annual %, const. 2005 prices)"].astype(float)
```

```python
df["GDP growth rate (annual %, const. 2005 prices)"]
```

```python
df["International trade: Balance (million US$)"].unique()
```

```python
df["International trade: Balance (million US$)"] = df["International trade: Balance (million US$)"].replace("...", np.nan)
```

```python
df["International trade: Balance (million US$)"] = df["International trade: Balance (million US$)"].astype(float)
```

Создадим бинарную переменную, которая показывает, является ли страна экспортером (trade = 1) или импортером (trade = 0)

```python
df['International trade: Balance (million US$)'].loc[(df['International trade: Balance (million US$)'] > 0)] = 1
```

```python
df['International trade: Balance (million US$)'].loc[(df['International trade: Balance (million US$)'] <= 0)] = 0
```

```python
df["Urban population growth rate (average annual %)"].unique()
```

```python
df["Urban population growth rate (average annual %)"] = df["Urban population growth rate (average annual %)"].replace("~0.0", "0.0")
```

```python
df["Urban population growth rate (average annual %)"] = df["Urban population growth rate (average annual %)"].astype(float)
```

```python
df["Population growth rate (average annual %)"].unique()
```

```python
df["Population growth rate (average annual %)"] = df["Population growth rate (average annual %)"].replace("-~0.0", "0.0")
```

```python
df["Population growth rate (average annual %)"] = df["Population growth rate (average annual %)"].replace("~0.0", "0.0")
```

```python
df["Population growth rate (average annual %)"] = df["Population growth rate (average annual %)"].replace("-~0.0", "0.0").astype(float)
```

```python
df["Fertility rate, total (live births per woman)"].unique()
```

```python
df["Fertility rate, total (live births per woman)"] = df["Fertility rate, total (live births per woman)"].replace("...", np.nan)
```

```python
df["Fertility rate, total (live births per woman)"] = df["Fertility rate, total (live births per woman)"].astype(float)
```

```python
d = {"Growth": df["GDP growth rate (annual %, const. 2005 prices)"], "Trade": df["International trade: Balance (million US$)"], "Fertility": df["Fertility rate, total (live births per woman)"], "Urban": df["Urban population growth rate (average annual %)"], "Pop_growth": df["Population growth rate (average annual %)"], "Pop_dens": df["Population density (per km2, 2017)"], "Industrialization": df["Economy: Industry (% of GVA)"], "Region": df["Region"]}
```

```python
f = pd.DataFrame(d)
f.head()
```

Удалим наблюдения, где в выбранных признаках есть пропущенные значения

```python
f = f.dropna()
```

```python
f.to_csv("prep_data2.csv")
```

```python
f["Growth"] = f["Growth"].replace(-99.0, np.median(f["Growth"]))
f["Fertility"] = f["Fertility"].replace(-99.0, np.median(f["Fertility"]))
f["Urban"] = f["Urban"].replace(-99.0, np.median(f["Urban"]))
f["Pop_growth"] = f["Pop_growth"].replace(-99.0, np.median(f["Pop_growth"]))
f["Industrialization"] = f["Industrialization"].replace(-99.0, np.median(f["Industrialization"]))
```

```python
f.to_csv("prep_median.csv")
```
